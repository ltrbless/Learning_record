```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x - 1) ** 2 + 2
x = np.linspace(-5, 7, 101)
y = f(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
```

```python
def df(theta):
    return 2 * (theta - 1)
```

$epsilon : \epsilon \ \ \ theta : \theta$

```python
theta = -4.0 # 初始值
learn_rate = 0.1 # 学习率
epoch = 100 # 学习次数
epsilon = 1e-8

theta_history = [theta] # 迭代历史

while epoch != 0 :
    epoch -= 1
    theta -= learn_rate * df(theta)
    theta_history.append(theta)
    if np.absolute(theta_history[-1] - theta_history[-2]) < epsilon :
        break
theta
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='b')
ax.plot(np.array(theta_history), f(np.array(theta_history)), '+', color='r')
plt.show()
```

```python
def gradient_descent(inital_theta, learn_rate, epoch=1000, epsilon=1e-8):
    theta_history = [inital_theta] # 迭代历史

    while epoch != 0 :
        epoch -= 1
        gradient = df(theta)
        theta -= learn_rate * gradient
        theta_history.append(theta)
        if np.absolute(theta_history[-1] - theta_history[-2]) < epsilon :
            break
            
    return theta_history[-1]
```

```python
def show(f):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x, y, color='b')
    ax.plot(np.array(theta_history), f(np.array(theta_history)), '+', color='r')
    plt.show()
```
