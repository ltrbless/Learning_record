```python
from Chapter04 import SimpleLinearRegression, metrics, model_selection, preprocessing
```

```python
from sklearn.datasets import load_boston
```

```python
dataset = load_boston()
```

```python
x = dataset.data[:, 5]
y = dataset.target
```

```python
x = x[y < 50]
y = y[y < 50]
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, test_ratio=0.2, random_state=666)
```

```python
LinearModel = SimpleLinearRegression.SimpleLinearRegression()
```

```python
LinearModel.fit(X_train, y_train)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
ax.plot(x, LinearModel.predict(x), color='r')
plt.show()
```

```python
metrics.r2_score(y_test, LinearModel.predict(X_test))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, LinearModel.predict(X_test))
```

```python
LinearModel.score(X_test, y_test)
```
