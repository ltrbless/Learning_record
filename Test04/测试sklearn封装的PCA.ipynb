```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
digits = datasets.load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 666)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
%time knn_clf.fit(X_train, y_train)
```

```python
knn_clf.score(X_test, y_test)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
X_pca = pca.transform(X)
```

```python
X_pca.shape
```

```python
for i in range(10):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1])
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(X_train)
X_train_reduce = pca.transform(X_train)
X_test_reduce = pca.transform(X_test)
```

```python
knn_clf2 = KNeighborsClassifier()
knn_clf2.fit(X_train_reduce, y_train)
knn_clf2.score(X_test_reduce, y_test)
```

```python
pca = PCA(0.95)
pca.fit(X_train)
X_train_reduce = pca.transform(X_train)
X_test_reduce = pca.transform(X_test)
knn_clf3 = KNeighborsClassifier()
knn_clf3.fit(X_train_reduce, y_train)
knn_clf3.score(X_test_reduce, y_test)
```
